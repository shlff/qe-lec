## 2.1 theory

Recall [09_mle_estimation_goodness_fit_us](https://github.com/jstac/inequality_dynamics/blob/master/firm_size_empirics/09_mle_estimation_goodness_fit_us.ipynb), we have

$$
\ln \frac{S_{t+1}}{S_{t}} = a + b \frac{1}{\ln S_t} \tag{1}
$$
where
- $S_t$ is the firm size at time $t$
- $a \sim N(\mu_{a}, \sigma^2_{a})$
- $b \sim N(\mu_{b}, \sigma^2_{b})$

Taking log of both sides of [(1)](#) yields

$$
\frac{S_{t+1}}{S_{t}} = \exp (a + b \frac{1}{\ln S_t})
$$
or equivalently,

$$
S_{t+1} = S_{t} \exp (a + b \frac{1}{\ln S_t})
$$
