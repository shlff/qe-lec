# 2 append and plots

## 2.1 append data sets

```python
df18 = df_us185[['year', 'noe', 'lognoe', 'fgrow_log']]
df17 = df_us175[['year', 'noe', 'lognoe', 'fgrow_log']]
df16 = df_us165[['year', 'noe', 'lognoe', 'fgrow_log']]
df15 = df_us155[['year', 'noe', 'lognoe', 'fgrow_log']]
df14 = df_us145[['year', 'noe', 'lognoe', 'fgrow_log']]
df13 = df_us135[['year', 'noe', 'lognoe', 'fgrow_log']]
```

```python
df_x1 = df18.append(df17, ignore_index=True)
df_x2 = df_x1.append(df16, ignore_index=True)
df_x3 = df_x2.append(df15, ignore_index=True)
df_x4 = df_x3.append(df14, ignore_index=True)
df_x5 = df_x4.append(df13, ignore_index=True)

df_x1
```

```python
df_x6 = df_x5[df_x5['fgrow_log'] != 0.0]
df_x7 = df_x6[df_x6['fgrow_log'] != np.inf]
df_x8 = df_x7[df_x7['fgrow_log'] != - np.inf]
df_x9 = df_x8[df_x8['noe'] != 0.0]
df_x10 = df_x9[df_x9['lognoe'] != 0.0]
df_x10
```

```python
df_x10.to_csv('data/USA19-13_top1.csv')
```
